```python
from sklearn.model_selection import train_test_split
```

```python
X = df[["Open","High","Low","Adj Close","Volume"]]
Y = df["Close"]
X
```

```python
Y
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y, test_size = 0.3, random_state = 42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train,Y_train)
```

```python
lm.coef_
```

```python
cdf=pd.DataFrame(lm.coef_,X.columns, columns = ['coef'])
cdf
```

```python
predict = lm.predict(X_test)
predict
```

```python
sns.scatterplot(x = predict, y = Y_test)
plt.xlabel("Predictions")
plt.title("Closing Price v/s Predictions accuracy depiction")
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
import math
```

```python
print("Mean Squared Error=", mean_squared_error(Y_test,predict))
print("Mean Absolute Error=", mean_absolute_error(Y_test,predict))
print("RMSE = ", math.sqrt(mean_squared_error(Y_test,predict)))
```
